```python
import pandas as pd
import os

# Load the dataset
file_path = "hp40.csv"
df = pd.read_csv(file_path)

# Extract the base name of the input file without extension
file_base_name = os.path.splitext(os.path.basename(file_path))[0]

# List of unique data types to filter and save separately
data_types = ["hr", "sleep", "br", "hrv", "spo2", "steps"]

# Save each data type into a separate CSV file with the original file name as a prefix
for dtype in data_types:
    df_filtered = df[df["data_type"] == dtype]
    if not df_filtered.empty:
        output_filename = f"{file_base_name}_{dtype}.csv"
        df_filtered.to_csv(output_filename, index=False)
        print(f"Saved {dtype} data to {output_filename}")
```

```python
import pandas as pd

# File path for the uploaded CSV
file_path = "blood_pressure_readings_ID40.csv"

# Load the dataset
df = pd.read_csv(file_path)

# Define the invalid value to be removed
invalid_value = -2147483648

# Filter out rows where any of the specified columns contain the invalid value
filtered_df = df[~((df["systolic"] == invalid_value) | 
                    (df["diastolic"] == invalid_value) | 
                    (df["pulse"] == invalid_value))]

# Define the output file path
output_file_path = "blood_pressure_readings_ID40_cleaned.csv"

# Save the cleaned data to a new CSV file
filtered_df.to_csv(output_file_path, index=False)

# Provide the output file path
output_file_path
```
